```python
telco_data = telecom_data.copy()
```

```python
telco_data['TotalCharges'] = pd.to_numeric(telco_data['TotalCharges'],errors='coerce' )
```

```python
telco_data.isna().sum()
```

```python
# finding percentage null in TotalCharges column
pct_null_totalcharges = 100*telco_data['TotalCharges'].isna().sum()/len(telco_data)
pct_null_totalcharges
```

```python
# since the missing data is minimum so we can ignore or drop that rows
telco_data.dropna(how = 'any',inplace=True)
```

```python
telco_data.isna().sum()
```

 creating tenure bins

```python
telco_data['tenure'].max()
```

```python
label = ["{0} - {1}".format(i,i+11) for i in range(1,72,12)]
telco_data['Tenure_group'] = pd.cut(telco_data['tenure'],range(1,80,12),right=False,labels=label)
```

```python
telco_data['Tenure_group'].head(10)
```

```python
telco_data['Tenure_group'].value_counts()
```

removing columns which are not required for processing

```python
# removing customer_id and tenure
telco_data.drop(columns=['customerID','tenure'],axis=1,inplace=True)
```

```python
telco_data
```

```python
plt.figure(figsize=(8,5))
for i,predictor in enumerate(telco_data.drop(columns=['Churn','TotalCharges','MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data=telco_data,x=predictor,hue='Churn')
```

```python
telco_data['Churn'] = np.where(telco_data['Churn']== 'Yes',1,0)
```

```python
telco_data['Churn'].head(5)
```

converting categorical data into dummy or indicator variables

```python
data_dummies = pd.get_dummies(telco_data)
data_dummies.head(5)
```

```python
plt.figure(figsize=(20,8))
data_dummies.corr()['Churn'].sort_values(ascending=False).plot(kind='bar')
```
